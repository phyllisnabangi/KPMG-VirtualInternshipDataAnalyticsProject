## Transactions

```python
# Load the dataset
transactions_data = pd.read_excel("KPMG_VI_New_raw_data_update_final.xlsx", "Transactions", skiprows=1)
transactions_data
```

```python
transactions_data.info()
```

```python
# checking for missing values
transactions_data.isna().sum()
```

```python
# checking for duplicates
transactions_data.duplicated().sum()
```

```python
# sorting the transactions data dataset by the date
transaction_sorted = transactions_data.sort_values(by=["transaction_date"], ascending=True)
```

```python
transaction_sorted
```

```python
complete_transactions = transactions_data['order_status'] == "Approved"
complete_transactions.sum()
```

```python
cancelled_transactions = transactions_data['order_status'] == "Cancelled"
cancelled_transactions.sum()
```

```python
# Visualizing for the Analysis of order status column
transactions_data['order_status'].value_counts().plot(kind='bar');

# plot label
plt.xlabel("order_status")
plt.ylabel("Count of orders made")
plt.title("Analysis of the order status")

plt.show()
```
